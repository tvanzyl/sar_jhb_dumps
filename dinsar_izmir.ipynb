```python
count=0
total=len(a)
pjobs = 8
!rm pexecute.sh
!touch pexecute.sh
for i in range(0,total):
    for j in range(i+1, min(i+1+overlap, total)):
        master = a[i]
        slave = a[j]
        print(master[39:47],slave[39:47])
        target_name = "{}_{}".format(master[39:47],slave[39:47])
        target = "{}{}_{}.dim".format(dinsar_dir,master[39:47],slave[39:47])
        c = !ls $dinsar_dir | grep $target_name
        print(c)        
        if len(c) == 0:            
            !echo "gpt graph/$graph -Pmaster=$master -Pslave=$slave -Ptarget=$target -PRg=$Rg -PAz=$Az -PcohRg=$cohRg -PcohAz=$cohAz -c 64G -q 9 &" >> pexecute.sh
            !echo "sleep 10s" >> pexecute.sh
            count += 1
        if (count-1)%pjobs==(pjobs-1):
            !echo "wait" >> pexecute.sh
            !bash pexecute.sh
            !rm pexecute.sh
            !touch pexecute.sh
if (count-1)%pjobs!=(pjobs-1):
    !echo "wait" >> pexecute.sh
    !bash pexecute.sh
!rm pexecute.sh
```
