```python
dinsar_path = "/data1/data/sar_rio_dumps/dinsar"
snaphu_path = "/data1/data/sar_rio_dumps/snaphu"
unwraped_path = "/data1/data/sar_rio_dumps/unwraped"
tc_path = "/data1/data/sar_rio_dumps/tc"
!mkdir $snaphu_path
!mkdir $unwraped_path
!mkdir $tc_path
```

```python
a = !ls $dinsar_path/*.dim
```

```python
a.sort()
```

Here we unwrap the interferogram using gpt.

```python
for i in range(len(a)):
    print(((((a[i]).split("/"))[-1]).split("."))[0])
    source_name = ((((a[i]).split("/"))[-1]).split("."))[0]
    source_path = "{}/{}.dim".format(dinsar_path, source_name)
    c = !ls $snaphu_path/ | grep $source_name
    print(c)
    if len(c) == 0:
        !gpt SnaphuExport -Ssource=$source_path -PtargetFolder=$snaphu_path
        %cd $snaphu_path/$source_name
        cmd_grep = !cat $snaphu_path/$source_name/snaphu.conf | grep "snaphu -f"
        cmd = cmd_grep[0].split("#")[1].strip()
        !$cmd
```

Here we inport the unwraped product back into snap to add the metadata correctly.

```python
b = !ls $snaphu_path
```

```python
b
```

```python
for i in range(len(b)):
    source_name = b[i]
    source_path1 = "{}/{}.dim".format(dinsar_path, source_name)    
    c = !ls $snaphu_path/$source_name/UnwPhase*.hdr
    source_path2 = c[0]    
    print(unwraped_path, source_name, source_path1, source_path2)
    d = !ls $unwraped_path | grep $source_name
    print(d)   
    if len(d) == 0:     
        !gpt SnaphuImport -t {"\""+unwraped_path+"/"+source_name+".dim"+"\""} {"\""+source_path1+"\""} {"\""+source_path2+"\""}
```

Next we need to ensure that the products are terrain corrected.

```python
b = !ls $unwraped_path/*.dim
```

```python
b
```

```python
for i in range(len(b)):
    source_path = b[i]
    source_name = source_path.split("/")[-1].split(".")[0]
    print(tc_path, source_name, source_path)
    d = !ls $tc_path | grep $source_name
    print(d)
    if len(d) == 0:
        !gpt Terrain-Correction -t {"\""+tc_path+"/"+source_name+".dim"+"\""} -PdemName="SRTM 1Sec HGT" {"\""+source_path+"\""}
```
