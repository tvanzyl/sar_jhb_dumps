```python
tc_dir = "/data1/data/sar_rio_dumps/tc"
dem_dir = "/data1/data/sar_rio_dumps/mintpy"
dinsar_dir = "/data1/data/sar_rio_dumps/dinsar"
mintpy_dir = "/data1/data/sar_rio_dumps/mintpy/interferograms"
!mkdir -p $dem_dir
!mkdir -p $mintpy_dir
```

```python
a = !ls $tc_dir/*.dim
print(a[0], a[1])
```

```python
a
```

```python
mintpy_subset = "mintpy_subset_rio.xml"
```

```python
for i in range(len(a)):    
    source = a[i]
    print(source)
    source_name = a[i].split("/")[-1].split(".")[0]
    coh_path = !ls {tc_dir}/{source_name}.data/coh*.hdr
    coh_band = coh_path[0].split("/")[-1].split(".")[0]
    print(coh_band)
    !mkdir $mintpy_dir/$source_name
    target_coh = "{}/{}/{}_coh_tc.dim".format(mintpy_dir,source_name,source_name)
    print(target_coh)
    unw_path = !ls {tc_dir}/{source_name}.data/Unw*.hdr
    unw_band = unw_path[0].split("/")[-1].split(".")[0]    
    print(unw_band)    
    target_unw = "{}/{}/{}_unw_tc.dim".format(mintpy_dir,source_name,source_name)    
    c = !ls $mintpy_dir/$source_name | grep *.dim
    print(c)
    if len(c) == 0:
        !gpt graph/$mintpy_subset -Psource=$source -Pcoh_band=$coh_band -Punw_band=$unw_band -Ptarget_unw=$target_unw -Ptarget_coh=$target_coh
```
