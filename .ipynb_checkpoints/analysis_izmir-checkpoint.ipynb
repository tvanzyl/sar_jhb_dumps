```python
def getRecIgram(ts, atr, exclude=None):
    range2phase = -4.*np.pi / float(atr['WAVELENGTH'])
    ts_data = ts * range2phase
    G_new = G.copy()
    ex_num = 0
    if not exclude is None:
        ex_num = G_new[:, exclude]
#         G_new[:, exclude] = 0.0        
        ts_data[exclude] = 0.        

    num_date, length, width = ts_data.shape
    ts_data = ts_data.reshape(num_date, -1)

    ifgram_est = np.dot(G_new, ts_data).reshape(G_new.shape[0], length, width)
    pha_data = np.array(ifgram_est, dtype=ts_data.dtype)

    pha_data[np.isnan(pha_data)] = 0.

    pha_data = pha_data[stack_obj.dropIfgram]
    rec_ifgram = pha_data.reshape(A.shape[0], -1)
    return rec_ifgram
```

```python
num_ifgram, num_pixel = ifgram.shape
ifgram_diff = ifgram - getRecIgram(ts, atr, 4)
inv_quality = np.abs(np.sum(np.exp(1j*ifgram_diff), axis=0)) / num_ifgram
```

```python
def moving_average(a, n=3):
    ret = np.cumsum(a, axis=0)
    ret[n:] = ret[n:] - ret[:-n]
    return ret[n - 1:] / n
```

```python
import statsmodels.api as sm
X_var = ['GDP']
dates = slice('2018-01-31','2019-12-31')
X = df_ts_M.loc[dates_prior, X_var]
y = df_ts_M.Velocity[dates_prior]
X = sm.add_constant(X)
# X['trend'] = range(1,len(X)+1)
# Note the difference in argument order
model = sm.OLS(y, X).fit(cov_type='HAC',cov_kwds={'maxlags':12})
# Print out the statistics
model.summary()
```

```python
import statsmodels.api as sm
X_var = ['GDP','Stringency']
dates = slice('2018-03-31','2020-09-30')
X = df_ts_Q.loc[dates, X_var]
y = df_ts_Q.Velocity[dates]
X = sm.add_constant(X)
# X['trend'] = range(1,len(X)+1)
# Note the difference in argument order
model = sm.OLS(y, X).fit(cov_type='HAC',cov_kwds={'maxlags':4})
X_all = df_ts_Q.loc[dates, X_var]
X_all.Stringency = 0.0
X_all = sm.add_constant(X_all)
# X_all['trend'] = range(1,len(X)+1)
predictions = model.predict(X_all) # make the predictions by the model
# Print out the statistics
model.summary()
```

```python
df_ts_Q["Prediction"] = predictions
df_ts_Q["Residual_Velocity"] = (df_ts_Q.Velocity-df_ts_Q.Prediction)
```
