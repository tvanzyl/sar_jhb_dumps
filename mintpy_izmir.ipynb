### Now we need to move the terrain correct files into the mintypy structure

The geoRegion for extraction of aligned extents, must be done to all three of the MintPy input products after terrain correction:
1. Unw
2. Coh
3. Dem

```python
#IZMIR
geoRegion = '"POLYGON((27.170101296927584 38.46831042986047,27.208038461234224 38.46831042986047,27.208038461234224 38.43766010905398,27.170101296927584 38.43766010905398,27.170101296927584 38.46831042986047))"'
```

The graph file used to subset out the coherence and unwrapped bands

```python
mintpy_subset = "mintpy_subset.xml"
```

Extract 
1. Coh
2. Unw

and then ensure we crop to the same extent.

```python
for i in range(len(a)):    
    source = a[i]
    print(source)
    source_name = a[i].split("/")[-1].split(".")[0]
    coh_path = !ls {tc_dir}/{source_name}.data/coh*.hdr
    coh_band = coh_path[0].split("/")[-1].split(".")[0]
    print(coh_band)
    !mkdir $mintpy_dir/$source_name
    target_coh = "{}/{}/{}_coh_tc.dim".format(mintpy_dir,source_name,source_name)
    print(target_coh)
    unw_path = !ls {tc_dir}/{source_name}.data/Unw*.hdr
    unw_band = unw_path[0].split("/")[-1].split(".")[0]    
    print(unw_band)    
    target_unw = "{}/{}/{}_unw_tc.dim".format(mintpy_dir,source_name,source_name)    
    c = !ls $mintpy_dir/$source_name | grep *.dim
    print(c)
    if len(c) == 0:
        !gpt graph/$mintpy_subset -Psource=$source -PgeoRegion=$geoRegion  -Pcoh_band=$coh_band -Punw_band=$unw_band -Ptarget_unw=$target_unw -Ptarget_coh=$target_coh
```

Next we need to add in the DEM, terrain correct and crop to the same extent

```python
b = !ls $dinsar_dir/*.dim
print(len(b))
b.sort()
```

```python
#'SRTM 1Sec HGT'
source = b[0]
!gpt AddElevation -Ssource=$source -PdemName='SRTM 1Sec HGT' -t $dem_dir/dem.dim
!gpt Terrain-Correction -t $dem_dir/dem_dinsar_tc.dim -Ssource=$dem_dir/dem.dim -PsourceBands="elevation" -PdemName="SRTM 1Sec HGT" 
!gpt Subset -t $dem_dir/dem_tc.dim -Ssource=$dem_dir/dem_dinsar_tc.dim -PgeoRegion=$geoRegion -PcopyMetadata=true -PsourceBands="elevation_VV"
```
