The graph file used to subset out the coherence and unwrapped bands

```python
# mintpy_subset = "mintpy_subset.xml"
mintpy_subset = "mintpy_subset_resample.xml"

width = 505
height= 290
```

Extract 
1. Coherence
2. Unwrapped Pahse
3. Intensity (For Later Calculations and Completness)

and then ensure we crop to the same extent.

```python
for i in range(len(a)):    
    source = a[i]
    print(source)
    source_name = a[i].split("/")[-1].split(".")[0]
    #Coherence
    coh_path = !ls {tc_dir}/{source_name}.data/coh*.hdr
    coh_band = coh_path[0].split("/")[-1].split(".")[0]
    print(coh_band)
    !mkdir $mintpy_dir/$source_name
    target_coh = "{}/{}/{}_coh_tc.dim".format(mintpy_dir,source_name,source_name)
    print(target_coh)
    #unwarpped Phase
    unw_path = !ls {tc_dir}/{source_name}.data/Unw*.hdr
    unw_band = unw_path[0].split("/")[-1].split(".")[0]    
    print(unw_band)    
    target_unw = "{}/{}/{}_unw_tc.dim".format(mintpy_dir,source_name,source_name)   
    print(target_unw)
    #Intensity
    int_path = !ls {tc_dir}/{source_name}.data/Intensity*_db.hdr
    int_band = int_path[0].split("/")[-1].split(".")[0]    
    print(int_band)    
    target_int = "{}/{}/{}_int_tc.dim".format(mintpy_dir,source_name,source_name)   
    print(target_int)
    
    c = !ls $mintpy_dir/$source_name | grep *.dim
    print(c)
    if len(c) == 0:
        !gpt graph/$mintpy_subset -Psource=$source -PgeoRegion=$geoRegion -Pcoh_band=$coh_band -Punw_band=$unw_band -Pint_band=$int_band -Ptarget_coh=$target_coh -Ptarget_unw=$target_unw -Ptarget_int=$target_int -Pwidth=$width -Pheight=$height
```
